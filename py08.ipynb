```python
import requests
from datetime import datetime, timedelta
from dotenv import load_dotenv
import json
import os

load_dotenv()
KEY = os.getenv("KEY")

today_date = datetime.today().strftime("%Y-%m-%d").replace("-", "")  # 오늘 날짜
yesterday_date = (datetime.today() - timedelta(1)).strftime("%Y-%m-%d").replace("-", "")  # 어제 날짜

url = f"http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key={KEY}&targetDt={yesterday_date}&itemPerPage=10&repNationCd=K"
req = requests.get(url)


data = req.json().get('boxOfficeResult', {}).get('dailyBoxOfficeList', [])
print(f"========== {today_date} 영화 순위 ==========")

for i, movie in enumerate(data,1):
    print(f"{i}위: {movie['movieNm']}")

```

```python
import requests as req
from bs4 import BeautifulSoup as bs
url = 'https://www.weather.go.kr/w/weather/forecast/short-term.do?stnId=159'
web =req.get(url)
soup = bs(web.content,'html.parser')
summary = soup.select('p.summary') #.select = 배열로 가져옴

for s in summary:
    print(s.text.replace('○','\n○'))
```

```python
table = soup.select('.table-col')[0]
data = table.select('tr')
header = data[0].select('th')
line1 = ['최저기온']+data[1].select('td')
line2 = ['최고기온']+data[2].select('td')
raw_data=[]
for h,l1,l2 in zip(header,line1,line2):
    try:
        out = h.text+":"+l1.text+"/"+l2.text
    except:
        out = h.text+":"+l1+"/"+l2
    finally:
        print(out)
        raw_data.append(out)
```

```python
import pandas as pd
# 데이터 피싱
parsed_data = []
for entry in raw_data[1:]:
    date, temps = entry.split(':')
    low_temp, high_temp = temps.split("/")
    parsed_data.append([date.strip(), low_temp.strip(), high_temp.strip()])

# 데이터프레임 생성
df = pd.DataFrame(parsed_data, columns=['날짜','최저기온','최고기온'])
# df.to_csv('weather.csv',index = False,encoding='utf-8')
df.to_csv('weather_x.csv',index = False,encoding='utf-8-sig')
df
```
